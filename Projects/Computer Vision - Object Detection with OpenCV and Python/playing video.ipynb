```python
#For playing 2 videos


import numpy as np
import cv2

names = ['cartoon1.mp4', 'earth.mp4'];
window_titles = ['first', 'second']


cap = [cv2.VideoCapture(i) for i in names]

frames = [None] * len(names);
#gray = [None] * len(names);
rgb = [None] * len(names);
ret = [None] * len(names);


while True:

    for i,c in enumerate(cap):
        if c is not None:
            ret[i], frames[i] = c.read();


    for i,f in enumerate(frames):
        if ret[i] is True:
            rgb[i] = cv2.cvtColor(f, cv2.COLOR_BGR2RGB)
            cv2.imshow(window_titles[i], rgb[i]);

    if cv2.waitKey(25) & 0xFF == ord('z'):
       break


for c in cap:
    if c is not None:
        c.release();

cv2.destroyAllWindows()
```
