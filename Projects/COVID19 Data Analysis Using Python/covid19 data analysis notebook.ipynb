### Task 2.3: Aggregating the rows by the country

```python
corona_dataset_agg = corona_dataset_csv.groupby("Country/Region").sum()
```

### Task3: Calculating a good measure

### task 3.1: caculating the first derivative of the curve

```python
corona_dataset_agg.loc["China"].diff().plot() #change in infection rate day by day
```

### task 3.2: find maxmimum infection rate for China

```python
corona_dataset_agg.loc["China"].diff().max() #in 1 day in 24 hours, there are 15136.0 new infected cases
```

```python
corona_dataset_agg.loc["Pakistan"].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_agg.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_agg.loc[c].diff().max())
corona_dataset_agg["max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_agg.head(10)
```

### Task 3.4: create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(corona_dataset_agg["max_infection_rates"])
```

```python
corona_data.head()
```
